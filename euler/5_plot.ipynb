```python
import math
from matplotlib import pyplot as plt

n = 100
results = list()
```

```python
# check if x is a prime number
def check_prime(x):
  for i in range(2, x):
    if x % i == 0:
      return False # if not return false
  return True # if prime return true
```

```python
def get_result(x):
    result = 1
    for i in range(2, x+1):
      if check_prime(i):
        result *= i**(math.floor(math.log(n)/math.log(i)))
    return result
```

```python
for i in range(1, n):
    results.append(math.log(get_result(i)))
```

```python
plt.plot(range(1, n), results)
plt.xlabel('n')
plt.ylabel('log(result)')
plt.rcParams['figure.figsize'] = [24, 16]
plt.rcParams['figure.dpi'] = 100
plt.show()
```
